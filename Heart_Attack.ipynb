```python
cholestrol = renamed_heart_df['Cholestrol(mg/dl)']
cholestrol
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Cholestrol(mg/dl)')
ax1.set_ylabel('Cholestrol')
ax1.boxplot(cholestrol)
plt.show()
```

```python
plt.savefig('Chol.png')
```

```python
quartiles = cholestrol.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of cholestrol is: {lowerq}")
print(f"The upper quartile of cholestrol is: {upperq}")
print(f"The interquartile range of cholestrol is: {iqr}")
print(f"The the median of cholestrol is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
